```python
# Importar la biblioteca random para la generación de números aleatorios
import random

# Importar la biblioteca statistics para cálculos estadísticos
import statistics

# Definición de la clase Cliente
class Cliente:
    def __init__(self, tiempo_llegada):
        self.tiempo_llegada = tiempo_llegada
        self.tiempo_servicio = random.expovariate(1 / E_S)  # Tiempo de servicio exponencial
        self.tiempo_cola = 0

# Definición de la clase Evento
class Evento:
    LLEGADA = 0
    INICIO_SERVICIO = 1
    FIN_SERVICIO = 2

# Definición de la clase SimuladorFacilidadServicio
class SimuladorFacilidadServicio:
    def __init__(self, s, E_A, E_S, n):
        # Inicialización de variables
        self.tiempo_actual = 0
        self.s = s  # Número de servidores
        self.E_A = E_A  # Media de los tiempos entre llegadas de clientes
        self.E_S = E_S  # Media de los tiempos de servicio de los clientes
        self.n = n  # Número de demoras para terminar la simulación
        self.clientes_en_cola = []  # Cola de clientes
        self.eventos = []  # Arreglo de eventos
        self.servidores = [False] * s  # Arreglo de estado de servidores
        self.tiempo_en_cola = 0
        self.total_clientes = 0
        self.total_servicio = 0
        self.total_demora_cola = 0
        self.total_clientes_cola = 0
        self.total_utilizacion_servidores = 0

    # Método para simular la llegada de un cliente
    def llegada_cliente(self):
        nuevo_cliente = Cliente(self.tiempo_actual)

        # Imprimir el evento de llegada
        print(f"Tiempo {self.tiempo_actual}: Llega cliente")
        self.total_clientes += 1
        print(f"Total de clientes hasta el momento: {self.total_clientes}")

        # Verificar si hay servidores libres para comenzar el servicio o agregar al cliente a la cola
        if self.servidores.count(False) > 0:
            self.clientes_en_cola.append(nuevo_cliente)
            print("Cliente inicia servicio")
            self.programar_inicio_servicio()
        else:
            print("Cliente se une a la cola")
            self.clientes_en_cola.append(nuevo_cliente)
            nuevo_cliente.tiempo_cola = self.tiempo_actual  # Guardar el tiempo de unión a la cola

        # Programar la próxima llegada de un cliente
        self.programar_llegada()

    # Método para simular el inicio del servicio
    def inicio_servicio(self):
        if self.clientes_en_cola:
            cliente_actual = self.clientes_en_cola.pop(0)
            if not len(self.clientes_en_cola) == 0:
                self.tiempo_en_cola += self.tiempo_actual - cliente_actual.tiempo_cola

            # Imprimir el tiempo de servicio del cliente actual
            print(f"Tiempo de servicio del cliente actual: {cliente_actual.tiempo_servicio}")
            self.total_servicio += cliente_actual.tiempo_servicio

            # Imprimir el tiempo total en cola
            if self.tiempo_en_cola > self.n:
                print(f"Tiempo en cola acumulado hasta el momento: 1000")
                return;
            else:
                print(f"Tiempo en cola acumulado hasta el momento: {self.tiempo_en_cola}")

            # Imprimir el inicio del servicio al cliente
            print(f"Tiempo {self.tiempo_actual}: Inicia servicio a cliente")

            # Marcar el servidor como ocupado y programar el fin del servicio
            indice_servidor_libre = self.servidores.index(False)  # Encuentra el primer servidor libre
            self.servidores[indice_servidor_libre] = True  # Marca el servidor como ocupado
            tiempo_fin_servicio = self.tiempo_actual + cliente_actual.tiempo_servicio
            self.eventos.append((tiempo_fin_servicio, Evento.FIN_SERVICIO))

    # Método para simular el fin del servicio
    def fin_del_servicio(self):
        print(f"Tiempo {self.tiempo_actual}: Finaliza servicio a cliente")
        indice_servidor_libre = self.servidores.index(True)  # Encuentra el primer servidor ocupado
        self.servidores[indice_servidor_libre] = False  # Marca el servidor como libre
        if self.clientes_en_cola:
            self.programar_inicio_servicio()  # Programar el inicio del servicio para el próximo cliente en cola

    # Método para programar la llegada de un nuevo cliente
    def programar_llegada(self):
        tiempo_llegada = self.tiempo_actual + random.expovariate(1 / self.E_A)
        self.eventos.append((tiempo_llegada, Evento.LLEGADA))

    # Método para programar el inicio del servicio
    def programar_inicio_servicio(self):
        self.eventos.append((self.tiempo_actual, Evento.INICIO_SERVICIO))

    # Método para simular el proceso
    def simular(self):
        total_demoras_cola_valores = []
        total_clientes_cola_valores = []
        total_utilizacion_servidores_valores = []

        # Realizar múltiples simulaciones
        i=0
        while( i < 1):
            i = 1
            # Reiniciar las variables para cada simulación
            self.tiempo_actual = 0
            self.clientes_en_cola = []
            self.eventos = []
            self.servidores = [False] * self.s
            self.tiempo_en_cola = 0
            self.total_clientes = 0
            self.total_servicio = 0

            # Programar el primer evento de llegada de un cliente
            self.programar_llegada()

            # Contadores para las métricas
            while self.tiempo_en_cola < self.n:
                if self.tiempo_en_cola >= self.n:
                    break
                # Ordenar eventos por tiempo y seleccionar el siguiente
                self.eventos.sort()
                tiempo_evento, tipo_evento = self.eventos.pop(0)

                # Actualizar tiempo actual
                self.tiempo_actual = tiempo_evento

                # Procesar el evento
                if tipo_evento == Evento.LLEGADA:
                    self.llegada_cliente()
                elif tipo_evento == Evento.INICIO_SERVICIO:
                    self.inicio_servicio()
                elif tipo_evento == Evento.FIN_SERVICIO:
                    self.fin_del_servicio()


# Parámetros de entrada
s = 5  # Número de servidores
E_A = 1  # Media de los tiempos entre llegadas de clientes
E_S = 4  # Media de los tiempos de servicio de los clientes
n = 1000  # Número de demoras para terminar la simulación

# Crear una instancia de la clase SimuladorFacilidadServicio y ejecutar la simulación
simulador = SimuladorFacilidadServicio(s, E_A, E_S, n)
simulador.simular()
```
